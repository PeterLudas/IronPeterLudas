```python
# checking NaN values

analysis.isnull().sum()
```

```python
# lowering characters in the dataframe including column names

for i in analysis:
    if analysis[i].dtype == 'object':
        analysis[i] = analysis[i].str.lower()

analysis.columns = analysis.columns.str.lower()
analysis.head()
```

```python
analysis.drop(columns = ['unnamed: 0', 'customer', 'effective to date', 'vehicle type', 'policy type', 'renew offer type', 'policy type', 'policy', 'vehicle type'], inplace = True)
analysis.head()
```
